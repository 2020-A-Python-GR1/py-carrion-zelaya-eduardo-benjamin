1) Crea un Dataframe de 10 registros y 6 columnas y consigue las 5 primeros y los 5 ultimos registros

```python
df10x6 = pd.DataFrame(np.random.randint(1, 10,60).reshape(10,6))
df10x6
```

```python
def imprimir_5_primeros_y_ultimos(dataframe):
    print('5 Primeros:\n')
    print(dataframe[:5])
    
    print('\n5 Ultimos:\n')
    print(dataframe[5:])
```

```python
imprimir_5_primeros_y_ultimos(df10x6)
```

4) Crear un Dataframe con 10 registros y 6 columnas y con una propiedad del Dataframe mostrar las columnas, con otro comando mostrar los valores.

```python
df10x6 = pd.DataFrame(np.random.randint(1, 10,60).reshape(10,6), columns=list('abcdef'))
print('Columnas:')
print(df10x6.columns)
print('\nMostrar columnas')
print(list(df10x6))

print('\nValores:')
print(df10x6.values)
```

6) Crear un Dataframe con 10 registros y 6 columnas y con una funcion del Dataframe transponer los datos

```python
df10x6 = pd.DataFrame(np.random.randint(1, 10,60).reshape(10,6), columns=list('abcdef'))
print('Inicial')
print(df10x6)
print('\n Transpuesto')
print(df10x6.transpose())
```
